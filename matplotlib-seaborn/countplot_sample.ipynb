```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df_train = pd.read_csv("train.csv")  # Titanic 数据
```

# 直观的查看每个feature所对应的feature

```python
cols = ['Survived', 'Sex', 'Pclass', 'SibSp', 'Parch', 'Embarked']
nr_rows = 2
nr_cols = 3 
fig, axs = plt.subplots(nr_rows, nr_cols , dpi = (200), figsize = (nr_cols*3.5,nr_rows*3))
sns.set_style("white")
for r in range(0,nr_rows):
    for c in range(0,nr_cols):
        i = r*nr_cols+c
        ax = axs[r][c]
        sns.countplot(df_train[cols[i]], hue = df_train['Survived'] , ax = ax , palette='Set1' )
        ax.set_title(cols[i])
        ax.legend(loc =1)
plt.tight_layout()
```
