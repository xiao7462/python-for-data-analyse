# the average total bill on each day

**agg** 函数会分组分后单独创建一个df

```python
df.groupby('day').agg('mean')
```

only inerested in [total_bill] column

```python
df.groupby('day')['total_bill'].agg('mean')
```

或者

```python
df.groupby('day')['total_bill'].mean()
```

# 平均餐费>20 ,根据days分类

**filter**函数会返回一个符合条件的subset

```python
df.groupby('day').filter(lambda x: x ['total_bill'].mean() > 20)
```

```python
import collections
collections.Counter((df.groupby('day').filter(lambda x: x ['total_bill'].mean() > 20)).day.values)
```

```python
collections.Counter(df.day.values)
```

filter后可以继续进行其他列的操作,例如求出 ['size']的平均值

```python
df.groupby('day').filter(
    lambda x : x['total_bill'].mean() > 20
)['size'].mean()
```

平均餐费<=20 ,根据days分类

```python
df.groupby('day').filter(
    lambda x : x['total_bill'].mean() <= 20
)['size'].mean()
```

# To make a new value for each row, use `pd.transform()`

将返回值重新填到df的每一行里

```python
df.groupby('day').transform(lambda x:x.mean())
```

只关注某个feature

```python
df.groupby('day')['total_bill'].transform(lambda x:x.mean())
```

## 可以添加一个新列到df中

```python
df['day_average'] =df.groupby('day')['total_bill'].transform( lambda x:x.mean());df
```
